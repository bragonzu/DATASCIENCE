```python
import pandas as pd

#Creamos 2 Dataframes, staff y estudiantes.
staff_df=pd.DataFrame([{'Name':'Kelly', 'Role':'Director of HR'},
                       {'Name':'Sally', 'Role':'Course liasion'},
                       {'Name':'James', 'Role':'Grader'}])

#Indexamos al personal por nombre
staff_df= staff_df.set_index('Name')

#Ahora crearemos un dataframe student
student_df= pd.DataFrame([{'Name':'James','School': 'Bussiness'},
                                {'Name':'Mikys','School': 'Law'},
                                {'Name':'Sally','School': 'Engineering'}])

#Indexamos los estudiantes por nombre.
student_df=student_df.set_index('Name')

#Imprimimos las salidas de los df
print(staff_df.head())
print(student_df.head())
```

```python
#Ahora haremos el merge de los 2 dataframes
#primero pasaremos del staff y luego del student
#How por defecto es interno lo que hacemos es sacarlo

pd.merge(staff_df,student_df,how='outer',left_index=True,right_index=True)
```

```python
#Vemos que aparece miky sin rol ni Kelly sin school
pd.merge(staff_df,student_df,how='inner',left_index=True,right_index=True)
```

```python
pd.merge(staff_df,student_df,how='left',left_index=True,right_index=True)
```

```python
pd.merge(staff_df,student_df,how='right',left_index=True,right_index=True)
```

```python
#Reseteamos los index
staff_df=staff_df.reset_index()
student_df=student_df.reset_index()

#Ahora fusionaremos con el parametro on.
pd.merge(staff_df,student_df,how='right', on='Name')
```

```python
staff_df=pd.DataFrame([{'Name':'Kelly','Role': 'Director of HR',
                        'Location':'State Street'},
                       {'Name':'Sally','Role': 'Course liasion',
                        'Location':'Washington Avenue'},
                        {'Name':'James','Role': 'Grader',
                         'Location': 'Washington Avenue'}])


student_df=pd.DataFrame([{'Name':'James','School': 'Bussiness',
                        'Location':'1024 Billigas Avenue'},
                         {'Name':'Mike','School': 'Bussiness',
                        'Location':'1024 Billigas Avenue'},
                         {'Name':'Sallys','School': 'Engineering',
                        'Location':'52352 Dire Avenue'}])

pd.merge(staff_df,student_df, how='left', on='Name')


```

```python
staff_df=pd.DataFrame([{'First Name':'Braham','Last Name':'Gons Zulo',
                        'Role':'Director of HR'},
                       {'First Name':'Sally','Last Name':'Brooks',
                        'Role':'Course liasion'},
                       {'First Name':'James','Last Name':'Zello',
                        'Role':'Grader'}])

student_df=pd.DataFrame([{'First Name':'James','Last Name':'Hammond',
                        'School':'Bussiness'},
                       {'First Name':'Mike','Last Name':'Psero',
                         'School':'Law'},
                       {'First Name':'Sally','Last Name':'Brooks',
                         'School':'Engineering'}])

#Indexación multiple usando on
pd.merge(staff_df, student_df, how='inner', on=['First Name','Last Name'])
```
