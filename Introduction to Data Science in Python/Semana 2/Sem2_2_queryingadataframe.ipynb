**QUERYING DATAFRAME**


```python
import pandas as pd

df=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/DATASETS/auto/auto-mpg.csv')
df.head()
```

```python
#REEMPLAZAMOS TODO LOS ATRIBUTOS A MAYUS
df.columns=[x.upper().strip() for x in df.columns] 
df.head(14)
```

```python
#CREAREMOS UNA MASCARA DE ADMISION PARA 'PROYECTAR' QUE REGISTRO CUMPLE
admit_mask=df['CYLINDERS']>=6 #LA MASCARA SOLO MUESTRA SI ES VERDADERO O FALSO LA COMPARACION
admit_mask
```

```python
#MOSTRAMOS LOS REGISTROS, PERO LOS QUE NO CUMPLEN CON LA CONDICION
#DE ARRIBA SE VUELVE NaN
df.where(admit_mask).head(14)
```

```python
#ELIMINAREMOS LOS REGISTROS QUE NO CUMPLEN LA CONDICION (NaN)
df.where(admit_mask).dropna().head()
```

```python
#MOSTRAMOS CYLINDERS  MAYOR A 6
df[df['CYLINDERS']>=6].head()
```
