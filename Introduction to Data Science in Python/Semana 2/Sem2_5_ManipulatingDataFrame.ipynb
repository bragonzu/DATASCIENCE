```python
import pandas as pd
df=pd.read_csv('/content/drive/MyDrive/Colab Notebooks/DATASETS/_DSWITHPYTHON/presidents.csv')
```

```python
df.head()
```

Creamos una nueva columna que es copia de presidente

```python
df['First']=df['President']
df.head()
```

En la columna first el primer espacio en blanco la cambiamos por una vacia


Ejemplo:

Input:Juan Medina Gomez 

Output:Juan

```python
df['First']=df['First'].replace("[ ].*","",regex=True)
df.head()
```

```python
df['Born']=df['Born'].str.extract("([\w]{3} [\w]{1,2}, [\w]{4})")
df['Born'].head()
```

```python
df['Born']=pd.to_datetime(df['Born'])
df['Born'].head()
```
